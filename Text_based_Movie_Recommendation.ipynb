```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/DataScience/RecommSys/popmovies.CSV', index_col=0)
print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
# Vectorizing the data
tfidf = TfidfVectorizer(min_df=4, max_df=0.7)

vectorized_data = tfidf.fit_transform(df['overview'])
print(vectorized_data.toarray())
```

```python
tfidf_df = pd.DataFrame(vectorized_data.toarray(), 
                        columns=tfidf.get_feature_names())
tfidf_df.index = df['title']
tfidf_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity_array = cosine_similarity(tfidf_df)
```

```python
cosine_sim_df = pd.DataFrame(cosine_similarity_array, index = tfidf_df.index, columns = tfidf_df.index)
print(cosine_sim_df)
```

```python
cosine_sim_df.loc['Bad Boys for Life'].sort_values(ascending=False).head()
```

```python
print(f"Supposed the user watched 'Bad Boys for Life', the most similar movies based on movie's overview are: {cosine_sim_df.loc['Bad Boys for Life'].sort_values(ascending=False).head()[1:].index}")
```
